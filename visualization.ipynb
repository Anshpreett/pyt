```python
fig = plt.figure(figsize=(10,4))
x = np.linspace(0, 10, 100)
y = np.sin(x)
z = np.cos(x)
plt.plot(x,y, linestyle='--', color='blue', linewidth=1, label='sin(x)')
plt.plot(x,z, linestyle='-.', color='red', linewidth=1, label='cos(x)')
plt.title('time series',loc='left',pad=10,fontdic={'fontsize':16,'color':})
plt.xlabel('time')
plt.ylabel('value')
plt.legend()
```

```python
fig = plt.figure(figsize=(10,4))
x = np.linspace(0, 10, 100)
y = np.sin(x)
z = np.cos(x)
plt.plot(x,y, linestyle='--', color='blue', linewidth=1, label='sin(x)')
plt.plot(x,z, linestyle='-.', color='red', linewidth=1, label='cos(x)')
plt.title('time series')
plt.legend()
plt.xlabel('time')
plt.ylabel('value')
plt.xlim(-1,12)
plt.ylim(-1.5,1.5)
```
