```python
fig = plt.figure(figsize=(10,4))
x = np.linspace(0, 10, 100)
y = np.sin(x)
z = np.cos(x)
plt.plot(x,y, linestyle='--', color='blue', linewidth=1, label='sin(x)')
plt.plot(x,z, linestyle='-.', color='red', linewidth=1, label='cos(x)')
plt.title('time series',loc='left',pad=10,fontdict={'fontsize':16,'color':'red'})
plt.xlabel('time')
plt.ylabel('value')
plt.legend()
```

```python
fig = plt.figure(figsize=(20,10))
x = np.linspace(0, 10, 100)
y = np.sin(x)
z = np.cos(x)
plt.plot(x,y, linestyle='--', color='blue', linewidth=1, label='sin(x)')
plt.plot(y,z, linestyle='-.', color='red', linewidth=1, label='cos(x)')
plt.plot(z,x, linestyle=':', color='green', linewidth=1, label='tan(x)')
plt.title('time series')
plt.legend()
plt.xlabel('time')
plt.ylabel('value')
plt.xlim(-1,10)
plt.ylim(-1.5,1.5)
```

```python
rainfall = [0.8,0.5,1.2,0.5,0.9,0.7,0.3,0.5,1.8,2,0.5,0.7]
month=['jan','feb','mar','apr','may','jun','jul','aug','sep','oct','nov','dec']
fig=plt.figure(figsize=(10,4))
plt.bar(month,rainfall,color='red',hatch='/\-.')
plt.title('rainfall in 2019')
plt.xlabel('Months')
plt.ylabel('Rainfall(mm)')
```

```python
plt.style.use('seaborn')
x = [1, 2, 3, 4]
colors = plt.get_cmap('rainbow')(np.linspace(0.2, 0.7, len(x)))
fig, ax = plt.subplots()
ax.pie(x, colors=colors, radius=3, center=(4.5, 4.5),hatch='/\-.',
       wedgeprops={"linewidth": 1, "edgecolor": "white"}, frame=True)
ax.set(xlim=(0, 8), xticks=np.arange(1, 10),
       ylim=(0, 8), yticks=np.arange(1, 10))
plt.title('data on a pie graph')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```
